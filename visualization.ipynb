## plot()

```python
bb=data.DataReader(name="MSFT",data_source="yahoo",start="2007-07-01",end="2008-12-31")
bb.head()
```

```python
bb.plot()  #high,low,open,close are so small in magnitude as compared to magintude of volume,hence their plot is lie along the line 0.0 in y-axis
```

```python
bb.plot(y="High")
```

```python
bb.plot(y=["High","Low","Open","Close"])
```

```python
bb["Open"].plot()
```
