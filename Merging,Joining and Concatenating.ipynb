## outer join

```python
w1.merge(w2,on="Customer ID",how="outer")  #full outer join
```

```python
a=w1.merge(w2,on="Customer ID",how="outer",indicator=True)
a
```

```python
a["_merge"].value_counts()
```

```python
b=a["_merge"].isin(["left_only","right_only"])
a[b]
```

## left join

```python
w2.merge(cust,how="left",right_on="ID",left_on="Customer ID") 
```

left_on and right_on are used generally when one column of one table has same value with anathor column of different table but those two columns have different name.It can be used in inner,outer,left,right join.
